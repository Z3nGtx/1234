```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding , LSTM,Dense
from sklearn.model_selection import train_test_split
import re
import numpy as np
```

```python
data=pd.read_csv("twitter_sentiments.csv",names=['id','loc','label','text'])
data
```

```python
data.shape
```

```python
data.dtypes
```

```python
data['text']=data['text'].astype(str)
```

```python
data
```

```python
def clean_text(text):
 text=text.lower()
 text=re.sub(r"[^a-zA-Z]+"," ",text)
 return text
```

```python
clean_text("hello friends! how are you?? welcome..62782!!!")
```

```python
data["text"]=data["text"].apply(clean_text)
```

```python
data
```

```python
comments=data["text"].tolist()
targets=data['label'].values
```

```python
pd.DataFrame(targets).value_counts()
```

```python
##generate training data
tokenizer=Tokenizer(num_words=5000)
tokenizer.fit_on_texts(comments)
sequences=tokenizer.texts_to_sequences(comments)
padded_sequences=pad_sequences(sequences,maxlen=200)
```

```python
padded_sequences.shape
```

```python
padded_sequences
```

```python
#output data preparation
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(targets)
from keras.utils import to_categorical
y_new=to_categorical(y)
```

```python
y_new
```

```python
x_train,x_test,y_train,y_test=train_test_split( padded_sequences,y_new,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# c.train the model
model=Sequential()
model.add(Embedding(5000,128, input_length=200))
model.add(LSTM(64))
model.add(Dense(4,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer="adam",metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=3,batch_size=32,validation_data=(x_test,y_test))
```

```python
new_comment=' I hate mango'
new_sequence=tokenizer.texts_to_sequences([clean_text(new_comment)])
padded_new_sequence=pad_sequences(new_sequence,maxlen=200)
prediction=model.predict(padded_new_sequence)[0]
```
